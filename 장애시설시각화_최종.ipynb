```python
import pandas as pd
import json
import folium
import googlemaps
import warnings
import folium
import numpy as np
warnings.simplefilter(action="ignore", category=FutureWarning)
```

```python
d_data = pd.read_csv("../동대문구 장애시설위치.csv")
s_data = pd.read_csv("../성북구 장애 시설 주소 최종.csv")
j_data = pd.read_csv("../중랑구 장애 시설 주소 최종.csv")
df = pd.read_csv("../FinalData.csv" , encoding="euc-kr")

df["장애인수대비시설수"] = df["장애인시설수"] / df["장애인구"] * 1000
```

```python
장애인시설합계=np.sum(df.장애인시설수)
장애인시설합계
장애인인구합계=np.sum(df.장애인구)
장애인인구합계
총장애인시설비율 = 장애인시설합계/장애인인구합계 *1000
총장애인시설비율
```

```python
dis_data = pd.pivot_table(df, index=["구"], values=["장애인수대비시설수"])
dis_data = dis_data.장애인수대비시설수.to_frame()
dis_data = dis_data.sort_values(by="장애인수대비시설수",ascending=True)
print(dis_data.head(3),"  ",dis_data.tail(3))
```

```python
geo_path = "C:\\Users\\SAS\\Desktop\\파이썬 프로젝트\\02. skorea_municipalities_geo_simple.json"
geo_data = json.load(open(geo_path, encoding="utf-8"))
```

```python
#장애인구대비 시설 부족 지역 : 중랑구, 성북구 ,동대문구
#장애인구대비 시설 충분 지역 : 강남구, 강동구 ,송파구 

#동대문구
loc = [d_data["위도"].mean(), d_data["경도"].mean()] # 위도(N), 경도(E)

map = folium.Map(location = loc, # 'OpenStreetMap'
                 zoom_start=14)
map.choropleth(geo_data=geo_data, data=dis_data, columns=[dis_data.index, "장애인수대비시설수"], fill_color="PuRd", key_on="feature.id")



data_size = len(d_data)

# 포인트 서클 추가

for i in range(data_size):

    folium.Circle(list(d_data.iloc[i][['위도', '경도']]),
                  radius = 1000,
                  color = '#2c9147',fill_color = '#2c9147').add_to(map)

# 포인트 마커 추가    
for i in range(data_size):

    folium.Marker(list(d_data.iloc[i][['위도', '경도']]),
                  popup = d_data.iloc[i]['주소'],
                 icon = folium.Icon(color='green')).add_to(map)
    

map
```
